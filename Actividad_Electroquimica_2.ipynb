```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
electrolitos = ["NaCl", "NaOH", "H2SO4", "H3PO4"]
concentraciones = [0.2, 0.1, 0.05, 0.01]

def voltaje_electrolisis(electrolito, concentracion):

  E0 = 1.23
  R = 8.314
  T = 298
  n = 2
  F = 96485

  Q_dict = {"NaCl": concentracion**2,
            "NaOH": concentracion,
            "H2SO4": concentracion,
            "H3PO4": concentracion}
  Q = Q_dict[electrolito]
  E = E0 - R * T / n / F * np.log(Q)

  return E

def eficiencia_electrolisis(electrolito, concentracion):

  n = concentracion
  R = 0.082
  T = 298
  P = 1
  V_teorico = n * R * T / P

  z = 2
  F = 96485
  I = 0.5
  t = 3600
  M = 2
  V_producido =  (I * t )/(z * F) * 100

  eficiencia = V_producido / V_teorico

  return eficiencia

def oxigeno_electrolisis(electrolito, concentracion):

  n = concentracion / 2
  R = 0.0821
  T = 298
  P = 1
  V = n * R * T / P
  return V

voltajes = []
eficiencias = []
oxigenos = []

for electrolito, concentracion in zip(electrolitos, concentraciones):

  voltaje = voltaje_electrolisis(electrolito, concentracion)
  voltajes.append(voltaje)
  eficiencia = eficiencia_electrolisis(electrolito, concentracion)
  eficiencias.append(eficiencia)
  oxigeno = oxigeno_electrolisis(electrolito, concentracion)
  oxigenos.append(oxigeno)

fig, ax1 = plt.subplots()

ax1.bar(electrolitos, voltajes, label="Voltaje (V)", color='blue')
ax1.set_xlabel("Electrolito")
ax1.set_ylabel("Voltaje (V)", color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

ax2 = ax1.twinx()
ax2.plot(electrolitos, eficiencias, color="red", marker="o", label="Eficiencia (%)")
ax2.set_ylabel("Eficiencia (%)", color="red")
ax2.tick_params(axis='y', labelcolor='red')


plt.title("Voltaje y eficiencia de la electrolisis del agua con diferentes electrolitos")
fig.tight_layout()
plt.legend()
plt.show()


plt.bar(electrolitos, oxigenos, label="Oxígeno (L)")

plt.ylim(0,10)
plt.xlabel("Electrolito")
plt.ylabel("Volumen (L)")
plt.title("Volumen del oxígeno producidos en la electrolisis del agua con diferentes electrolitos")
plt.legend()
plt.show()

```
