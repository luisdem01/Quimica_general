```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

##1. Generar datos usando drawdata para obtener:

###(a) Ley de Boyle

```python
drawdata.draw_scatter()
```

```python
boyle = pd.read_csv('data.csv')

x = boyle['x']
y = boyle['y']

coeficientes = np.polyfit(x, y, 2)
regresion_lineal = np.poly1d(coeficientes)

valores_x = np.linspace(min(x), max(x), 500)
plt.plot(valores_x, regresion_lineal(valores_x), color='red')

plt.scatter(boyle['x'], boyle['y'])
plt.title('Grafico Ley de Boyle')
plt.xlabel('Volumen (L)')
plt.ylabel('Presion (atm)')

plt.show()

print(f"La ecuacion de la curva es: {round(coeficientes[0],3)}x^2 + ({round(coeficientes[1],3)})x + {round(coeficientes[2],3)}")
```

###(b) Ley de Charles

```python
drawdata.draw_scatter()
```

```python
charles = pd.read_csv('data (1).csv')

x = charles['x']
y = charles['y']

coeficientes = np.polyfit(x, y, 1)
regresion_lineal = np.poly1d(coeficientes)

valores_x = np.linspace(min(x), max(x), 500)
plt.plot(valores_x, regresion_lineal(valores_x), color='red')

plt.scatter(charles['x'], charles['y'])
plt.title('Grafico Ley de Charles')
plt.xlabel('Temperatura (K)')
plt.ylabel('Volumen (L)')

plt.show()

print(f"La ecuacion de la recta es: {round(coeficientes[0],3)}x + {round(coeficientes[1],3)}")
```
